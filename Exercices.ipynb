**Exercice 2 :** Mapping

```python
import pandas as pd
data={'Name':['A','B','C','D'],'Age':[42,54,20,63]}
df=pd.DataFrame(data,index=['value1','value2','value3','value4'])
df
```

```python
score={'A':24,'B':30,'C':36,'D':42}
# Conversion du premier caractère du nom en majuscule pour que les noms soient identiques. 
n=df["Name"].str.capitalize()
n
```

Ecrire un programme Pandas qui sert à ajouter la variable de score à l'ensemble de données df à l'aide de la méthode map(). 

```python
#solution
my_series = score={'A':24,'B':30,'C':36,'D':42},
    index=["1","2","3","4"])

altered_series=my_series.map(lambda x: str(x)+".00")

print("Initial Series:")
print(my_series,"\n")

print("Altered Series:")
print(altered_series,"\n")
```
